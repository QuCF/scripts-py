```python
# ---------------------------------------------------------------------------------------------------
# --- Consider the Lotka-Volterra prey-predator model:
# --- d x / d t =  y = Fx ---
# --- d y / d t = -x + mu*(1-x^2)y = Fy ---
# ---------------------------------------------------------------------------------------------------
mix.reload_module(mix)

# Nt = 1200
Nt = 1200
t_array = np.linspace(0, 12, Nt)
dt = np.diff(t_array)[0]

print(dt)

# x0, y0 = 0.5, 0.5  # growth with saturation
# x0, y0 = 2.1, 2.1 # oscillations (stable)
x0, y0 = -1.0, 1.0 # 
mu = 0.5

Fx = lambda x, y:  y
Fy = lambda x, y: -x + mu*(1-x**2)*y
```

```python
# ---------------------------------------------------------------------------------------------------
# --- Direct solver:
# ---------------------------------------------------------------------------------------------------
mix.reload_module(mix)

class DirectSolver__:
    xt_ = None
    yt_ = None
    Nt_ = None
    dt_ = None
    Fx_ = None
    Fy_ = None
    t_ = None
    def __init__(self, x0, y0, Fx, Fy, t):
        self.t_ = np.array(t)
        self.Nt_ = len(self.t_)
        self.dt_ = np.diff(self.t_)[0]
        self.Fx_ = Fx
        self.Fy_ = Fy
        self.xt_ = np.zeros(self.Nt_, dtype=float)
        self.yt_ = np.zeros(self.Nt_, dtype=float)
        self.xt_[0] = x0
        self.yt_[0] = y0
        return

    def solve(self):
        for it in range(Nt-1):
            self.xt_[it+1] = self.xt_[it] + self.dt_ * self.Fx_(self.xt_[it], self.yt_[it])
            self.yt_[it+1] = self.yt_[it] + self.dt_ * self.Fy_(self.xt_[it], self.yt_[it])
        return
        

ds = DirectSolver__(x0, y0, Fx, Fy, t_array)    
ds.solve()
```

```python
# --- Plot results from the direct solver ---
plt.close()
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(ds.t_, ds.xt_, color='b', linewidth = 2, linestyle='-', label="x")
ax.plot(ds.t_, ds.yt_, color='r', linewidth = 2, linestyle='-', label="y")
plt.xlabel('$t$')
plt.ylabel("$x,y$")
ax.legend()
plt.grid(True)
plt.show()
```

```python
# ---------------------------------------------------------------------------------------------------
# --- KvN solver:
# ---------------------------------------------------------------------------------------------------
class Liouville__:
    psi_ = None
    nx_, ny_ = None, None
    Nx_, Ny_ = None, None # the number of points to describe a single variable;
    t_ = None
    delta_x_, delta_y_ = None, None
    x0_, y0_ = None, None
    x_, y_ = None, None
    dx_, dy_ = None, None

    def __init__(self, x0, y0, t):
        # time grid
        self.t_ = np.array(t)
    
        # grids for the variables:
        self.nx_, self.ny_ = 7, 7
        self.Nx_ = 1 << self.nx_
        self.Ny_ = 1 << self.ny_
        max_r = 4.0
        self.x_ = np.array(np.linspace(-max_r, max_r, self.Nx_), dtype='float32')
        self.y_ = np.array(np.linspace(-max_r, max_r, self.Ny_), dtype='float32')
        self.dx_ = np.diff(self.x_)[0]
        self.dy_ = np.diff(self.y_)[0]

        dt = np.diff(self.t_)[0]
        cx = max_r * dt / self.dx_
        cy = (max_r + mu*(1-max_r**2)*max_r) * dt / self.dy_
        print("Courant numbers: {:0.3e}, {:0.3e}: ".format(cx, cy))
    
        # state vector:
        self.psi_ = np.zeros((Nt, self.Nx_, self.Ny_), dtype='float32')
    
        # Initial state parameters:
        self.delta_x_ = 0.05
        self.delta_y_ = 0.05
        self.x0_ = x0
        self.y0_ = y0
        
        coef_gauss = 1./(2.*np.pi * self.delta_x_ * self.delta_y_)
        for iy in range(self.Ny_):
            y1 = self.y_[iy]
            for ix in range(self.Nx_):
                x1 = self.x_[ix]
                temp  = np.exp(-(x1 - self.x0_)**2/(2.*self.delta_x_**2))
                temp *= np.exp(-(y1 - self.y0_)**2/(2.*self.delta_y_**2)) * temp
                self.psi_[0, ix, iy] = temp
        self.psi_ = np.sqrt(coef_gauss * self.psi_)
    
# -----------------------------------
st = time.time()

kvn = Liouville__(x0, y0, t_array)

et = time.time()
elapsed_time = et - st
print('Execution time:', elapsed_time, 'seconds')
```

```python
# Compute the KvN:
# @njit
# def solve_Liouville(psi, mu, Nt, dt, x, y, Nx, Ny, dx, dy):  
#     # we use Dirichlet boundary conditions:
#     for it in range(Nt-1):
#         for iy in range(1, Ny-1):
#             y1 = y[iy]
#             for ix in range(1, Nx-1):
#                 x1 = x[ix]
#                 Fx1   = y1
#                 Fx1_p = y1
#                 Fx1_m = y1

#                 coef_y = mu * (1 - x1*x1)
#                 Fy1   = -x1 + coef_y * y1
#                 Fy1_p = -x1 + coef_y * y[iy+1]
#                 Fy1_m = -x1 + coef_y * y[iy-1]

#                 sign_x   = Fx1   / np.abs(Fx1)
#                 sign_x_p = Fx1_p / np.abs(Fx1_p)
#                 sign_x_m = Fx1_m / np.abs(Fx1_m)

#                 sign_y   = Fy1   / np.abs(Fy1)
#                 sign_y_p = Fy1_p / np.abs(Fy1_p)
#                 sign_y_m = Fy1_m / np.abs(Fy1_m)
                
#                 psi1 = psi[it, ix, iy]

#                 temp_x = 0
#                 if sign_x > 0:
#                     temp_x += Fx1 * psi[it, ix, iy]
#                 else:
#                     temp_x -= Fx1 * psi[it, ix, iy]
#                 if sign_x_p < 0:
#                     temp_x += Fx1_p * psi[it, ix+1, iy]
#                 if sign_x_m > 0:
#                     temp_x -= Fx1_m * psi[it, ix-1, iy]
                
#                 temp_y = 0
#                 if sign_y > 0:
#                     temp_y += Fy1 * psi[it, ix, iy]
#                 else:
#                     temp_y -= Fy1 * psi[it, ix, iy]
#                 if sign_y_p < 0:
#                     temp_y += Fy1_p * psi[it, ix, iy+1]
#                 if sign_y_m > 0:
#                     temp_y -= Fy1_m * psi[it, ix, iy-1]

#                 psi[it+1, ix, iy] = psi1 - dt/dx * temp_x - dt/dy * temp_y
#     return


@njit
def solve_Liouville_v2(psi, mu, Nt, dt, x, y, Nx, Ny, dx, dy):  
    # we use Dirichlet boundary conditions:
    for it in range(Nt-1):
        for iy in range(1, Ny-1):
            y1 = y[iy]
            for ix in range(1, Nx-1):
                x1 = x[ix]
                Fx1   = y1
                Fy1   = -x1 + mu * (1 - x1*x1) * y1
                sign_x   = Fx1   / np.abs(Fx1)
                sign_y   = Fy1   / np.abs(Fy1)

                psi1 = psi[it, ix, iy]

                temp_x = 0
                if sign_x > 0:
                    temp_x += Fx1 * (psi1 - psi[it, ix-1, iy])
                else:
                    temp_x += Fx1 * (psi[it, ix+1, iy] - psi1)
                    
                temp_y = 0
                if sign_y > 0:
                    temp_y += Fy1 * (psi1 - psi[it, ix, iy-1])
                else:
                    temp_y += Fy1 * (psi[it, ix, iy+1] - psi1)

                psi[it+1, ix, iy] = psi1 - dt/dx * temp_x - dt/dy * temp_y
    return


# def solve_Liouville_exp(psi, mu, Nt, dt, x, y, Nx, Ny, dx, dy): 
#     return




# -------------------------------------------
st = time.time()

# solve_Liouville(
solve_Liouville_v2(
    kvn.psi_[:,:,::1], 
    mu, 
    len(kvn.t_), np.diff(kvn.t_)[0],
    kvn.x_,  kvn.y_,
    kvn.Nx_, kvn.Ny_,
    kvn.dx_, kvn.dy_
)

et = time.time()
elapsed_time = et - st
print('Execution time:', elapsed_time, 'seconds')
```
