```python
# -------------------------------------------------
# --- Modeling the logistic map ---
# -------------------------------------------------
N = 60
x = np.zeros(N)

r = 3.2
x[0] = 0.6


for i in range(1,N):
    xp = x[i-1]
#     x[i] = r * xp * (1 - xp)
    x[i] = xp**2
axis_n = np.array(range(N))
    
    
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(axis_n, x, color='b', linewidth = 2, linestyle='-', label="x")
plt.xlabel('i-th iteration')
plt.ylabel("$x_i$")
ax.legend()
plt.grid(True)
plt.show()    
```
