```python
# find angles for the matrix structure:
d = 1./np.sqrt(2.)

a1 = 0.1
a2 = 0.3
a3 = 0.2
a4 = 0.4

th1 = 2. * np.arccos(a1/d**2) 
th2 = 2. * np.arccos(a2/d**2)
th3 = 2. * np.arccos(a3/d**2)
th4 = 2. * np.arccos(a4/d**2)

print("th1 {:0.10f}".format(th1))
print("th2 {:0.10f}".format(th2))
print("th3 {:0.10f}".format(th3))
print("th4 {:0.10f}".format(th4))
```
