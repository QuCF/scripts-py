```python
# -----------------------------------------------
# --- Option 1: Compute all singular values ---
# -----------------------------------------------
flag_zero_diff = True

ix, iv = 9, 5

# --- Chosen matrix ---
str_diff = ""
if flag_zero_diff:
    str_diff = "-diff-zero"
AA = data_.cl_["{:d}{:d}{:s}".format(ix, iv, str_diff)]['A'].get_dense_matrix()

# --- Compute singular values ---
_, svs, _ = np.linalg.svd(AA)

min_sv = np.min(np.abs(svs))
max_sv = np.max(np.abs(svs))
print("max. sv:\t {:0.3e}".format(max_sv))
print("min. sv:\t {:0.3e}".format(min_sv))
print("cond. number:\t {:0.3e}".format(max_sv/min_sv))
```
