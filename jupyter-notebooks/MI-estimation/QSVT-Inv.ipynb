```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
import os
print(os.getpid())
%cd ../../

colors_ = ["blue", "red", "green", "gray", "black"]

import pylib.mix as mix
import pylib.qucf_read as qucf_r
import pylib.measurement as mse
```

```python
path_qucf_ = "../QuCF/simulations/MI/Inv"
name_qucf_pr_   = "MI"
name_qucf_file_ = name_qucf_pr_ + "_OUTPUT.hdf5"

path_save_plots_ = "./tools/QSVT-angles/inversion/saved-plots/"

for i in range(100):
    plt.close()
```

```python
# ------------------------------------------------------------------
# --- Create a reference matrix ---
# ------------------------------------------------------------------
def prepare_ref(flag_full):
    # --- Print initial parameters for BE ---
    print("nx\t {:d}".format(nx_))
    print("kappa\t {:0.1f}".format(kappa_goal_))

    if flag_full:
        norm_coef = 1.0
        xmax = 1.00
    else:
        norm_coef = 0.99
        xmax = 1.00

    # --- x grid ---
    Nx_ = 1 << nx_
    x_grid_1 = np.linspace(-xmax, -1.0/kappa_goal_, Nx_//2)
    x_grid_2 = np.linspace(1.0/kappa_goal_, xmax, Nx_//2)
    x_ = np.concatenate((x_grid_1, x_grid_2))

    # --- function to inverse ---
    y_diag_ = ( 1. - np.exp(-(5*kappa_goal_*x_)**2) ) / x_
    y_diag_ /= kappa_goal_
    y_diag_ *= norm_coef

    # --- Rearranged grid ---
    y_diag_re_ = np.zeros(Nx_)
    for ii in range(Nx_//2):
        y_diag_re_[ii] = y_diag_[Nx_//2-1 - ii]
    for ii in range(Nx_//2):
        y_diag_re_[Nx_//2 + ii] = y_diag_[Nx_-1 - ii]

    # --- matrix whose singular values encode the function ---
    A_ = np.diag(y_diag_)

    # --- reference inverse function ---
    y_inv_ref_ = 1. / y_diag_
    y_inv_ref_ /= kappa_goal_

    # --- Plotting the intial function ---
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(x_, y_diag_, color='b', marker = ".", linewidth = 1, linestyle='-')
    plt.xlabel('$x$')
    plt.title("y-diag")
    plt.grid(True)
    plt.show()

    # --- Plotting the initial function ---
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(x_, y_inv_ref_, marker = ".", color='b', linewidth = 1, linestyle='-')
    plt.xlabel('$x$')
    plt.title("y-inv / kappa")
    plt.grid(True)
    plt.show()
    return x_, Nx_, y_diag_, y_diag_re_, A_, y_inv_ref_
# --------------------------------------------------------------
kappa_goal_ = 1.0e6
nx_ = 7

flag_full_ = False

x_, Nx_, y_diag_, y_diag_re_, A_, y_inv_ref_ = prepare_ref(flag_full = flag_full_)
```

```python
# ----------------------------------------------------------
# --- Compute the condition number ---
# ----------------------------------------------------------
from numpy import linalg as LA

evals, evecs = LA.eig(A_)
kappa_ = np.max(np.abs(evals)) / np.min(np.abs(evals))
print("nx: \t{:d}".format(nx_))
print("cond. number: {:0.1e}".format(kappa_))
```

```python
# -------------------------------------------------
# --- Matrix inversion ---
# -------------------------------------------------
mix.reload_module(mse)
def compare_with_qucf(flag_full, flag_adv, flag_save):
    # --- Reading QC data ---
    km, ke = mix.get_order_base10(kappa_goal_)
    str_k = "{:0.1f}e{:d}".format(km, ke)
    pr_name = "MI_k{:s}".format(str_k)

    str_full = ""
    path_res = path_qucf_ + "/results"
    if flag_adv:
        pr_name += "_ADV"
    if flag_full:
        str_full = "_FULL"
        path_res = path_qucf_ + "/results" + str_full + "/"

    om = mse.MeasOracle__(True)
    om.path_  = path_res
    om.pname_ = pr_name
    om.open()

    # --- QSVT parameter ---
    om.read_qsvt()
    kappa_qc = om.dd_["qsvt_mi"]["par"]
    if np.abs(kappa_qc - kappa_goal_) > 1e-4:
        print()
        print("<<< ERROR: a wrong QSVT parameter.>>>")
        print()
    print()
    print("QSVT kappa: {:0.1e}".format(kappa_qc))

    # --- QuCF data ---
    om.set_zero_ancillae_work_states(0)
    qc_data_all = om.get_var_x({}, "rs")

    y_inv_qc_ = qc_data_all.imag  # correct global phase;

    y_inv_qc_ *= 2**(nx_/2.) / om.dd_["qsvt_mi"]["rescaling_factor"]
    if y_inv_qc_[0] > 0:
        y_inv_qc_ *= -1 # correct global phase;

    # --- Relative error ---
    norm_diff = y_inv_ref_ - y_inv_qc_
    max_rel_err = np.max(np.abs(norm_diff))
    log_norm_diff = np.log10(np.abs(norm_diff))
    print()
    print("max. relative err: {:0.3e}".format(max_rel_err))

    # --- Plotting: comparison ---
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(
        x_, y_inv_ref_, 
        color='b', linewidth = 2, linestyle='-',  
        label = "ref"
    )
    ax.plot(
        x_, y_inv_qc_,
        color='r', linewidth = 2, linestyle='--',  
        label = "QC"
    )
    plt.xlabel('$x$')
    plt.title("inv-function")
    ax.legend()
    plt.grid(True)
    plt.show()

    # --- Plotting: error ---
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(x_, log_norm_diff, color='b', linewidth = 2, linestyle='-')
    plt.xlabel('$x$')
    plt.title("log10 norm. (ref - reco)")
    plt.grid(True)
    plt.show()

    if flag_save:
        # mix.save_dat_plot_1d_file(
        #     path_save_plots_ + "/QSVT_Dir_ref_k{:s}{:s}.dat".format(str_k, str_full),
        #     x_, y_inv_ref_
        # )
        # mix.save_dat_plot_1d_file(
        #     path_save_plots_ + "/QSVT_Dir_reco_k{:s}{:s}.dat".format(str_k, str_full),
        #     x_, y_inv_qc_
        # )
        mix.save_dat_plot_1d_file(
            path_save_plots_ + "/QSVT_Dir_err_x_k{:s}{:s}.dat".format(str_k, str_full), 
            x_, log_norm_diff
        )
    return
# ----------------------------------------------------------
compare_with_qucf(
    flag_full = flag_full_,
    flag_adv  = False,
    flag_save = False
)
```
